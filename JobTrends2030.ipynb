```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\kirti\Desktop\it_jobs_2030.csv',encoding= 'unicode_escape')
```

```python
print (df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Get the top 10 domains by count
top_10_domains = df['Domain'].value_counts().nlargest(10)

# Create the bar chart
sns.set()
plt.figure(figsize=(15,6))
sns.barplot(x=top_10_domains.index, y=top_10_domains.values)
plt.title("Top 10 Domains by Count")
plt.xlabel("Domain")
plt.ylabel("Count")
plt.show()

```

```python
plt.figure(figsize=(10,6))
plt.pie(df['Domain'].value_counts(), labels = df['Domain'].unique(), autopct='%1.1f%%')
plt.title("Domain Distribution")
plt.show()
```

```python
# Create a pivot table
pivot_data = pd.pivot_table(df, values='Job Title', index='Domain', aggfunc='count', fill_value=0)

sns.set()
plt.figure(figsize=(10,6))
sns.heatmap(pivot_data, annot=True, cmap="Blues")
plt.title("Domain vs. Job Title Heatmap")
plt.xlabel("Job Title")
plt.ylabel("Domain")
plt.show()
```
